# Seaborn

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
mydata={'Name' :['Ram','Sam','Joe','Asha'],
        'Age' :[23,22,26,47],
        'Salary':[1200,50000,56000,54000],
        'Exp':[2,1,3,10]
       }
```

```python
df=pd.DataFrame(mydata)
df.head()
```

Histgram

```python
plt.figure(figsize=(6,5))
sns.histplot(df['Salary'],kde = True,bins = 2)
plt.title('Distibution of Salary')
plt.show()
```

1.There is an negative skew, lower salary,below average
2.The average Salary is 30000
3.There is no outerliers
4.Majority salary is blw 28000 to 56000 

Correlation Matrix(heatmap)

```python
#step1: filter the numerical value
ndf=df.select_dtypes(include =['number'])
ndf.head()
```

```python
#step2: heatmap
plt.figure(figsize=(6,5))
sns.heatmap(ndf.corr(), cmap='coolwarm', annot=True)
plt.title('Correlation between sal,age,exp')
plt.show()
```

1.Dark area more correlated
2.light coloured area are less correlated

Box plot

```python
plt.figure(figsize=(5,6))
sns.boxplot(x= df['Age'])
plt.title('Age Distribution')
plt.show()
```

1.the average age is 24
2.47 is outlier in  the data

```python
mydata1={'Name' :['Ram','Sam','Joe','Asha'],
        'Age' :[23,22,26,47],
        'Salary':[1200,50000,56000,54000],
        'Exp':[2,1,3,10],
        'G':['M','F','M','M']
       }
df1=pd.DataFrame(mydata1)
```

```python
plt.figure(figsize=(6,5))
sns.countplot(x=df1['Exp'],palette='pastel',hue=df1['G'])
plt.title('Counts of exp')
plt.show()
```
